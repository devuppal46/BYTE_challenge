# 1 PACKAGES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import copy

%matplotlib inline
```

# 2 Import dataset ,identify target & feature

```python

House_DF=pd.read_csv('housing_price_dataset.csv')
X=House_DF[['SquareFeet', 'Bedrooms', 'Bathrooms', 'YearBuilt']] #features
y=House_DF['Price'] #target
```

# 3 MAKING TRAINING AND TESTING DATA SETS

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=101)
```

# 4 TRAIN  LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
regression= LinearRegression()
regression.fit(X_train,y_train)
print(regression.coef_)
print(regression.intercept_)
```

# 5 TEST/ PREDICT  FROM MODEL

```python
Y_PREDICT= regression.predict(X_test)
```

# 6 EVALUATE PERFORMANCE  OF MODEL

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,Y_PREDICT)
print("COEFFICIENT OF PERFORMANCE :",score)
```

# 7 PLOT  AND VISUALISE

```python
# plt.scatter(y_test,Y_PREDICT)

plt.figure(figsize=(10,6))
plt.scatter(y_test,Y_PREDICT,alpha=0.3)
plt.title('HOUSE PRICE PREDICTION')

plt.show()
```
